```python
import yfinance as yf
import mplfinance as mpf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
nvda=yf.Ticker('NVDA').history('5y')
```

```python
mpf.plot(nvda)
```

```python
data=pd.read_csv('NVDA.csv')
```

```python
data.info()
```

```python
data['Date'] = pd.to_datetime(data['Date'])  # change datetime format 
data=data.set_index('Date')          # set date as index
```

```python
mpf.plot(data)
```

```python
data2022 = data['2022']
```

```python
mpf.plot(data['2021'])
```

```python
mpf.plot(data['2020'], type='line')
```

```python
mpf.plot(data['2021'], type='candle')
```

```python
mpf.plot(data['2020'], type='renko')
```

```python
mpf.plot(data2022, type='pnf')
```

```python
mpf.plot(data['2021'], type='candle', volume=True)
```

```python
mpf.plot(data['2020-07':'2020-12'], type='candle', 
         volume=True, mav=(20,5))
```

```python
mpf.plot(data['2020-07':'2020-12'], type='candle', 
         volume=True, mav=(20,5),title = 'Nivdia 2020',
        tight_layout=True, figratio=(10,5))
```

```python
mpf.plot(data['2020-07':'2020-12'], type='candle', volume=True, 
         mav=(20,5), style='yahoo')
```

```python
mpf.available_styles()
```

```python
mpf.plot(data['2020-07':'2020-12'], type='candle', volume=True, 
         mav=(20,5), style='yahoo', 
         savefig='img.png' )
```

```python
mpf.plot(data, type='candle', volume=True, 
         title="03, 2021", 
         figratio=(14,6),
         mav=(5,10,30), tight_layout=True, 
         style='mike', savefig='Nivdia.png')
```
